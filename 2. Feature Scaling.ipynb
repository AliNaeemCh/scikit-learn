# Feature Scaling

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('datasets/drawndata1.csv')
```

```python
df.head(5)
```

```python
plt.scatter(df['x'], df['y'])
```

```python
X = df[ ['x', 'y'] ]
X.head(5)
```

### StandardScaler
- Sensitive to outliers.

- Formula:
#### $$x_{scaled} = \frac{x-\mu}{\sigma}$$

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
# Fit method in sklearn allows a model to learn particular parameters from the dataset passed.
X_fitted_ss = ss.fit(X)
# The learned parameters would be different from model to model, let's extract the parameters of StandardScaler
print('The mean extracted/learned by StandardScaler is: ', X_fitted_ss.mean_)
print('The variance extracted/learned by StandardScaler is: ', X_fitted_ss.var_)
```

```python
# After learning the parameters, it's time to apply them to the dataset. That is done by the transform command:
X_scaled_ss = X_fitted_ss.transform(X)
```

```python
plt.scatter(X_scaled_ss[:,0], X_scaled_ss[:,1])
plt.title('Output of StandardScaler')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
# You can also perform these two operations: fitting and transforming using a single command 'fit_transform':
X_scaled_SS = ss.fit_transform(X)
```

```python
plt.scatter(X_scaled_SS[:,0], X_scaled_SS[:,1])
plt.title('Output of StandardScaler')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

### MinMaxScaler
- Sensitive to outliers.

- Formula:
#### $$x_{scaled} = \frac{x-x_{min}}{x_{max}-x_{min}}$$

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
# By default, the range of values is from 0 to 1, inclusive.
# You can also modify the range of scaled features as:
# minMax = MinMaxScaler(feature_range=(1,2))
X_scaled_mm = mm.fit_transform(X)  # Outputs as numpy array
```

```python
plt.scatter(X_scaled_mm[:,0], X_scaled_mm[:,1])
plt.title('Output of MinMax Scaler')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```
