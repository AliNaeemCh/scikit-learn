# One Hot Encoder

```python
import pandas as pd
```

```python
dict = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000],
      'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'],
      'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
```

```python
df = pd.DataFrame(dict)
df.head(5)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
# Parameters explanation:

# handle_unknown='ignore' 
# It ignores any unknown category, which wasn't present during fitting, while transforming the dataset.

# sparse_output=False
# It makes the matrix with 1s and 0s explicitly defined rather than the sparse output which has non-zero values with
# their indices.
```

```python
city_encoded = ohe.fit_transform(df[['city']])
city_encoded
```

```python
X_encoded = pd.concat([ df[['sales', 'size']], city_encoded], axis='columns')
X_encoded
```
